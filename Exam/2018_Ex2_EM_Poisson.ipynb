(2) EM algorithm

Consider N observations $x_n, n=1,...,N$, assumed to be i.i.d. from a mixture of Poisson distribution:

$$p(x_n|\pi,\lambda) = \sum^K_{k=1} \pi_k Poisson(x_n|\lambda_k)$$

Represent the model using latent variables and derive the E step of the EM algorithm, which could be used to learn the maximum likelihod estimates for the parameters $\pi = (\pi_1,...,\pi_K)^T$ and $\lambda=(\lambda_1,...,\lambda_K)^T$

The EM algorithm is an iterative method for finding maximum likelihood estimates of parameters in statistical models involving latent variables. In the case of a mixture of Poisson distributions, we can introduce a set of latent variables z_nk, where z_nk = 1 if observation x_n was generated by the k-th Poisson distribution and z_nk = 0 otherwise.

The complete data likelihood can be written as:

p(x,z|pi,lambda) = product(n=1 to N) product(k=1 to K) [pi_k * Poisson(x_n|lambda_k)]^z_nk

In the E step of the EM algorithm, we compute the expected value of the complete data log-likelihood with respect to the conditional distribution of the latent variables given the observed data and current parameter estimates. This can be written as:

Q(pi,lambda|pi(t),lambda(t)) = E_z[log p(x,z|pi,lambda)|x,pi(t),lambda(t)]

where pi^(t) and lambda^(t) are the current estimates of the parameters.

To compute this expectation, we need to evaluate the posterior probabilities of the latent variables given the observed data and current parameter estimates. These probabilities are given by:

p(z_nk=1|x_n,pi(t),lambda(t)) = pi_k^(t) * Poisson(x_n|lambda_k^(t)) / sum(j=1 to K) pi_j^(t) * Poisson(x_n|lambda_j^(t))

Substituting these probabilities into the expression for Q(pi,lambda|pi(t),lambda(t)), we obtain:

Q(pi,lambda|pi(t),lambda(t)) = sum(n=1 to N) sum(k=1 to K) p(z_nk=1|x_n,pi(t),lambda(t)) * [log pi_k + log Poisson(x_n|lambda_k)]

This is the expression that needs to be maximized in the M step of the EM algorithm to obtain updated estimates of the parameters pi and lambda.
